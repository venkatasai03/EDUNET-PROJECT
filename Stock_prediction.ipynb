# **Understanding data**

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM,Dropout
from tensorflow.keras.callbacks import ModelCheckpoint,EarlyStopping

import warnings
warnings.filterwarnings("ignore")

import os
os.environ["TF_CPP_MIN_LOG_LEVEL"]="3"
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
def check_df(dataframe, head=5):
    print("#################### Shape #################### ")
    print(dataframe.shape)
    print("#################### Types #################### ")
    print(dataframe.dtypes)
    print("#################### Head #################### ")
    print(dataframe.head(head))
    print("#################### Tail #################### ")
    print(dataframe.dtypes)
    print("#################### NA #################### ")
    print(dataframe.isnull().sum())
    print("#################### Quantiles #################### ")
    # Convert numerical columns to numeric data type before calculating quantiles
    numeric_df = dataframe.select_dtypes(include=np.number)
    print(numeric_df.quantile([0, 0.5, 0.50, 0.95, 0.99, 1]).T)
```

# **Preparing the data**

```python
tesla_data=data[["Date","Close"]]
```

```python
tesla_data.head()
```

```python
print("Min. Tarih:",tesla_data["Date"].min())
print("Max. Tarih:",tesla_data["Date"].max())
```

```python
tesla_data.index=tesla_data["Date"]
```

```python
tesla_data
```

```python
tesla_data.drop("Date",axis=1,inplace=True)
```

```python
tesla_data
```

```python
result_data=tesla_data.copy
```

```python
plt.figure(figsize=(12,6))
plt.plot(tesla_data["Close"],color="blue");
plt.ylabel("Stock Price")
plt.title("Tesla Stock Price")
plt.xlabel("Time")
plt.show()
```

```python
tesla_data=tesla_data.values
```

```python
tesla_data[0:5]
```

```python
tesla_data=tesla_data.astype("float32")
```

```python
def split_data(dataframe,test_size):
    pos=int(round(len(dataframe)*(1-test_size)))
    train=dataframe[:pos]
    test=dataframe[pos:]
    return train,test,pos
```

```python
train,test,pos=split_data(tesla_data,0.20)
```

```python
print(train.shape,test.shape)
```

```python
scaler_train=MinMaxScaler(feature_range=(0,1))
```

```python
train=scaler_train.fit_transform(train)
```

```python
scaler_test=MinMaxScaler(feature_range=(0,1))
```

```python
test=scaler_test.fit_transform(test)
```

```python
train[0:5]
```

```python
test[0:5]
```

```python
def create_features(data,lookback):
    X,Y=[],[]
    for i in range(lookback,len(data)):
        X.append(data[i-lookback:i,0])
        Y.append(data[i,0])
    return np.array(X),np.array(Y)
```

```python
lookback=20
```

# **Train dataset**

```python
X_train,y_train=create_features(train,lookback)
```

# **Test dataset**

```python
X_test,y_test=create_features(test,lookback)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
X_train[0:5]
```

```python
y_test[0:5]
```

```python
X_train=np.reshape(X_train,(X_train.shape[0],1,X_train.shape[1]))
```

```python
X_test=np.reshape(X_test,(X_test.shape[0],1,X_test.shape[1]))
```

```python
y_train=y_train.reshape(-1,1)
```

```python
y_test=y_test.reshape(-1,1)
```

```python
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

# **Modeling**

```python
model=Sequential()
model.add(LSTM(units=50,
              activation="relu",
              input_shape=(X_train.shape[1],lookback)))
model.add(Dropout(0.2))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(loss="mean_squared_error",optimizer="adam")
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint

callbacks = [
    EarlyStopping(monitor="val_loss", patience=3, verbose=1, mode="min"),
    ModelCheckpoint(filepath="mymodel.keras", monitor="val_loss", mode="min",
                    save_best_only=True, save_weights_only=False, verbose=1)
]
```

```python
history = model.fit(x=X_train,
                    y=y_train,
                    epochs=100,
                    batch_size=20,
                    validation_data=(X_test,y_test),
                    callbacks=callbacks,
                    shuffle=False)
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,2)
plt.plot(history.history["loss"],label="Training Loss")
plt.plot(history.history["val_loss"],label="Validation Loss")
plt.legend(loc="upper right")
plt.xlabel("Epoch",fontsize=16)
plt.ylabel("Loss",fontsize=16)
plt.ylim([0,max(plt.ylim())])
plt.title("Training and Validation Loss",fontsize=16)
plt.show()
```

#**Evaluation**


```python
loss=model.evaluate(X_test,y_test,batch_size=20)
```

```python
print("\nTest loss:%.1f%%"%(100.0*loss))
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
train_predict=scaler_train.inverse_transform(train_predict)
test_predict=scaler_test.inverse_transform(test_predict)
```

```python
y_train=scaler_train.inverse_transform(y_train)
y_test=scaler_test.inverse_transform(y_test)
```

```python
train_rmse=np.sqrt(mean_squared_error(y_train,train_predict))
```

```python
test_rmse=np.sqrt(mean_squared_error(y_test,test_predict,))
```

```python
print(f"Train RMSE:{train_rmse}")
print(f"Test RMSE:{test_rmse}")
```
